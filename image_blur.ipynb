```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread("/content/output1/image_0_1731.jpg")
```

```python
image.shape
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
gray.shape
```

```python
blurred1 = blurred = cv2.blur(image, (9,9))
```

```python
gaus_blurred = cv2.GaussianBlur(image, (5, 5), 0)
```

```python
cv2_imshow(gaus_blurred)
```

```python
cv2_imshow(blurred1)
```

```python
median_blurred = cv2.medianBlur(image, 5)
```

```python
cv2_imshow(median_blurred)
```
